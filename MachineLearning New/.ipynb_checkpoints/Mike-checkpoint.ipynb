```python
list_lengths = np.array(FifaDB['Height'])
empty_list = []
for x in list_lengths:
    splitted_values = x.split("'")
    feet = (float(x[0])) + (float(x[2])/10)
    centimeters = feet * 30.48
    empty_list.append(round(centimeters))

print(empty_list)
```

```python
list_wages = FifaDB['Wage']
wages = []
for i in list_wages:
    hoi = i.strip('K').strip('€')
    wage = int(hoi) * 1000
    wages.append(wage)
print(wages)
```

```python
list_value = FifaDB['Value']
list_new_values = []

for v in list_value[1:10]:
    if v[-1] == 'M':
        strip_value = v.strip('M').strip('€')
        new_value = strip_value * 1000000
        list_new_values.append(new_value)
print(list_new_values)
    
```
